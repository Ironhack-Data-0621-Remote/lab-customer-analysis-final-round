Pick any two queries from the queries that we have written so far and execute them in Python. The objective is to retrieve the data into a dataframe.

```python
query = '''
        SELECT *, rank() over (order by amount desc) as 'rank'
        FROM bank.loan
        order by status
        '''
data = pd.read_sql_query(query, engine)
data.head()
```

```python
query = '''     
        SELECT round(avg(amount),2) as "Avg Amount", round(avg(payments),2) as "Avg Payment", status
        FROM bank.loan
        GROUP BY status
        ORDER BY status;
        ''' 
data = pd.read_sql_query(query, engine)
data.head()
```
